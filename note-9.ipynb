4が出るまでの数字を表示する。４じゃない間は繰り返しなさいが下のプログラム。いきなり４が出ることもある。rand_numは４じゃない場合は繰り返す。（１でも１００でも4以外ならなんでもいい）

問題１０６

```python
import random
while True:
  rand_num = random.randint(1,6)
  print(rand_num)
  if rand_num == 4:
    break
  print('.')
```

Trueだから無限ループのこと。Breakをかくと無限ループを止める。（今回は４が出れば止めるという意味）無限ループの中でピリオドで表示される。Breakがあると次に表示されたものを実行せずに終わるから最後のピリオドは表示されない。

問題１０７

```python
import random
while True:
  rand_num = random.randint(1,6)
  print(rand_num)
  if rand_num == 4:
    continue
  print('.')
```

continueがあるとピリオドは表示しない。continueがあるとその先を無視して止まる。

```python
for x in range(1,21):
    if x % 5 == 0:
     print('Buzz')
    else:
     print(x)
```

```python
for x in range(1,21):
    if x % 3 == 0:
     print('Fizz')
    if x % 5 == 0:
     print('Buzz')
    else:
     print(x)
```

問題１１４

```python
for x in range(1,21):
    if x % 3 == 0:
        if x % 5 == 0:
            print('Fizz Buzz')
        else:
            print('Fizz')
    elif x % 5 == 0:
        print('Buzz')
    else:
        print(x)
```

```python
for x in range (1.21):
    if x % 3 == 0 and x % 5 != 0:
        print('Fizz')
    if x % 3 != 0 and x % 5 == 0:
        print('Buzz')
    if x % 3 == 0 and x % 5 == 0:
        print('Fizz Buzz')
    if x % 3 != 0 and x % 5 != 0:
        print(x)
```
